**Import important libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Read in the csv file as a dataframe called df**

```python
df = pd.read_csv('child-mortality-around-the-world.csv')
```

**Check info of df**

```python
df.info()
```

**Check columns of df**

```python
df.columns
```

```python
df.rename(columns = {'Child mortality rate - Sex: all - Age: 0-4 - Variant: estimates':'Child mortality'},
          inplace = True) 
```

**Check for missing values in df**

```python
df.isnull().sum()
```

**Drop column containing missing values**

```python
df.drop(columns=["Code"], inplace = True)
```

**Set up the plot**

```python
fig, ax = plt.subplots(figsize=(35, 20))

africa = df[df['Entity'] == 'Africa (UN)'].reset_index(drop=True)
asia = df[df['Entity'] == 'Asia (UN)'].reset_index(drop=True)
europe = df[df['Entity'] == 'Europe (UN)'].reset_index(drop=True)
nor_america = df[df['Entity'] == 'Northern America (UN)'].reset_index(drop=True)
sou_america = df[df['Entity'] == 'Latin America and the Caribbean (UN)'].reset_index(drop=True)
oceania = df[df['Entity'] == 'Oceania (UN)'].reset_index(drop=True)
world = df[df['Entity'] == 'World'].reset_index(drop=True)


ax = sns.lineplot(x='Year', y='Child mortality', data = africa)
ax = sns.lineplot(x='Year', y='Child mortality', data = asia)
ax = sns.lineplot(x='Year', y='Child mortality', data = europe)
ax = sns.lineplot(x='Year', y='Child mortality', data = nor_america)
ax = sns.lineplot(x='Year', y='Child mortality', data = sou_america)
ax = sns.lineplot(x='Year', y='Child mortality', data = oceania)
ax = sns.lineplot(x='Year', y='Child mortality', data = world, linewidth = 3)


plt.text(1970, 23.5, "Africa", fontsize=25, color='tab:blue', fontweight='light', rotation=-20)
plt.text(1952, 23.5, "Asia", fontsize=25, color='tab:orange', fontweight='light', rotation=-30)
plt.text(1965, 4, "Europe", fontsize=25, color='tab:green', fontweight='light', rotation=-7)
plt.text(1953, 3.7, "North America", fontsize=25, color='tab:red', fontweight='light', rotation=-5)
plt.text(2000, 8, "World", fontsize=28, color='tab:pink', fontweight='semibold', rotation=-15)
plt.text(1980, 7.2, "South America", fontsize=25, color='tab:purple', fontweight='light', rotation=-20)
plt.text(1975, 5, "Oceania", fontsize=25, color='tab:brown', fontweight='light', rotation=-10)

plt.text(1947, 32, "32 %", fontsize=20, color='tab:blue', fontweight='light')
plt.text(2022, 6.5, "6.6 %", fontsize=20, color='tab:blue', fontweight='light')

plt.text(1947, 25, "25 %", fontsize=20, color='tab:orange', fontweight='light')
plt.text(2022, 2.7, "2.7 %", fontsize=20, color='tab:orange', fontweight='light')

plt.text(1946.5, 22.5, "22.5 %", fontsize=20, color='tab:pink', fontweight='light')
plt.text(2022, 3.7, "3.7 %", fontsize=20, color='tab:pink', fontweight='light')

plt.text(1947, 20.8, "21 %", fontsize=20, color='tab:purple', fontweight='light')
plt.text(2022, 1.3, "1.6 %", fontsize=20, color='tab:purple', fontweight='light')

plt.text(1947, 9.3, "9 %", fontsize=20, color='tab:brown', fontweight='light')
plt.text(2022, 1.9, "1.9 %", fontsize=20, color='tab:brown', fontweight='light')

plt.text(1947, 8.5, "8.7%", fontsize=20, color='tab:green', fontweight='light')
plt.text(2022, -0.1, "0.4 %", fontsize=20, color='tab:green', fontweight='light')

plt.text(1947, 3.8, "3.8 %", fontsize=20, color='tab:red', fontweight='light')
plt.text(2022, 0.5, "0.6 %", fontsize=20, color='tab:red', fontweight='light')


plt.xlabel("Year", fontsize=28)
plt.ylabel("Child Mortality (%)", fontsize=25)
ax.set_yticklabels([])
ax.set_xlim(1946, 2026)
ax.tick_params(axis='x', which='major', labelsize=25)

plt.title('% of children dying before five years old', fontsize=30)

plt.text(2015, -4, "Source: Our World in Data", fontsize=18, color='black', fontweight='light')

plt.savefig("child_mortality.png", dpi = 300)
plt.show()
```
